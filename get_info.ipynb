```python

def svs(input_file):# pars SVS file
    #input_file
    #print(len(elefile))
    inp=input_file.replace("   ", " ")
    inp=inp.replace("  ", " ")
    inp = inp.replace(" ", " ")
    inp=inp.strip()
    inp= inp.split(" ")
    mpiq=inp.pop(0)
    inp = list(map(int, inp))
    return inp
def param(param_file):# pars PARAM file
    par=param_file.replace("    ", " ")
    par=par.replace("   ", " ") 
    par=par.replace("  ", " ")
    par=par.replace(" ", " ")
    par=par.strip()
    par= par.split(" ")
    par = list(map(float, par))# perevod v float
    return par

def get_info(gns_file):
    t = 0
    svsList=list()# arr for svs names
    paramGNS=[] # arr for parame
    timeSvs=[] # arr for time param
    dataGNS=[]# arr svs = param
    dataEpoch=[]# arr svs = epoch
    filename =   os.path.basename(gns_file)
    file = open (filename)
    allLines = file.readlines()
    filelinecount = len(allLines)
    file.seek(0) # peremestit ukozatel v nachal
    for ln in range(filelinecount) :#3700 strok все строки
        line  = file.readline()
        if ln>3 and ln % 2 ==0: # не четные строки         
        #print(elefile)
            if len(line) > 3:##       SVS ION
                S = svs(line)  #func
#                print(S)
            svsList.append(S)
        if (ln % 2) and (ln > 4) != 0: # не четные строки
        
            if len(line) > 3:
                p = param(line)
                tt = np.full(len(p),t/3600)# massiv vremeni zapolnyaet kol element, znachenie

            paramGNS.append(p)
            timeSvs.append(tt)# dla vremeni 
            t += 30  # vremya uvelich      
    #print(len(timeSvs))
    sVs = np.array(svsList,dtype=object)
    PGNS = np.array(paramGNS,dtype=object)
    tS = np.array(timeSvs,dtype=object)
    
    for i in range(len(PGNS)):
        dictionary = dict(zip(sVs[i],PGNS[i]))
        dataGNS.append(dictionary)
    for i in range(len(PGNS)):
        dictionary_b = dict(zip(sVs[i],tS[i]))
        dataEpoch.append(dictionary_b)    
    df = pd.DataFrame(dataGNS)
    dt = pd.DataFrame(dataEpoch)
    df = df.reindex(sorted(df.columns), axis=1)# SORTIROVK COLUMNS
    dt = dt.reindex(sorted(dt.columns), axis=1)# SORTIROVK COLUMNS
    #df.insert(loc=0, column="epoch", value=tS)  
    
    return df, dt
#Зделать функцию по проверки и удалению лишних таблиц спутников для равенства таблиц
def sravnenie(d1,d2):# Проверяет датафрэймы на количество столбцов
    arr1 =[]
    arr2 =[]
    if d1.shape[1]!=d2.shape[1]:
        for i in d1:
            arr1.append(i)
        for i in d2:
            arr2.append(i)    
    result = list(set(arr1) ^ set(arr2))
    return result
######################################################################
#azi,azit= get_info('ABD60613a.azi')
ele,eleT = get_info('ABD60613a.ele')
print(eleT)
#mp,mpT = get_info('ABD60613a.mp1')
#io,ioT = get_info('ABD60613a.ion')
#print(eleT.iloc[900:926])
#ele29 = eleT[29]
#ele29.dropna(inplace=True)
#print(ele29)
#gg = pd.DataFrame()
#gg['azi']=azi[29]
#gg['io']=io[29]
#gg['T']=ioT[29]
#gg['ele']=ele[29]

#gg.dropna()
#import plotly.express as px
#import plotly

#aaa = sravnenie(mp,io)
#print(aaa)
#print(azi.shape[1])
#print(azi.shape[1])
#print(ele.shape[1])
#print(eleT.shape[1])
#for i in range(len(p)):
#    a_dictionary = dict(zip(s[i],p[i]))
#    dataGNS.append(a_dictionary)
   
print("ok")    
```
